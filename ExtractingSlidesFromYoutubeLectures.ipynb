```python
import os
import shutil

# Specify the source folder and the destination folder
source_folder = '/content/Lecture-Video-to-PDF'
destination_folder = '/content'

# List all files and subdirectories in the source folder
contents = os.listdir(source_folder)

# Move each item from the source folder to the destination folder
for item in contents:
    source_item = os.path.join(source_folder, item)
    destination_item = os.path.join(destination_folder, item)
    shutil.move(source_item, destination_item)

# Delete the now empty source folder
shutil.rmtree(source_folder)
```

```python
from pytube import Playlist
from moviepy.editor import VideoFileClip, concatenate_videoclips

# Replace 'your_playlist_url' with the URL of your YouTube playlist
playlist_url = 'https://www.youtube.com/playlist?list=PLrv_b_Wg7BhRKpMzB88MMrw0IFPyz_iDs'

# Create a Playlist object
playlist = Playlist(playlist_url)

# Set the resolution to 360p
resolution = '360p'

# Initialize a list to store video clips
video_clips = []

# Loop through each video in the playlist and download
for video in playlist.videos:
    # Stream filter for the desired resolution
    stream = video.streams.filter(res=resolution, only_video=True).first()

    if stream:
        # Download the video without audio
        stream.download(output_path='/content/mydownload')
        # Create a VideoFileClip object from the downloaded video
        video_clip = VideoFileClip(f'/content/mydownload/{video.title}.mp4')
        video_clips.append(video_clip)
    else:
        print(f"No {resolution} stream found for {video.title}")

# Reverse the list of video clips
video_clips.reverse()

# Concatenate the video clips
final_video = concatenate_videoclips(video_clips)

# Set the output resolution (optional)
final_video = final_video.resize(resolution)

# Export the final video
final_video.write_videofile('merged_video.mp4', codec='libx264')

print("Download and merge completed.")
```
